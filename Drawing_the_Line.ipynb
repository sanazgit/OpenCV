```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_size = 600
image = np.ones((image_size, image_size), dtype=np.uint8) * 200

# Number of divisions on each side
divisions = 16

# step size
step_size = image_size // (divisions + 1)

for i in range(1, divisions + 1):
    # Top to right
    cv2.line(image, (i * step_size, 0), (image_size - 1, i * step_size), 0)
    # Right to bottom
    cv2.line(image, (image_size - 1, i * step_size), ((divisions - i + 1) * step_size, image_size - 1), 0)
    # Bottom to left
    cv2.line(image, ((divisions - i + 1) * step_size, image_size - 1), (0, (divisions - i + 1) * step_size), 0)
    # Left to top
    cv2.line(image, (0, (divisions - i + 1) * step_size), (i * step_size, 0), 0)

plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
