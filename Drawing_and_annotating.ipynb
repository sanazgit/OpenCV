```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image= np.zeros((500,500,3), np.uint8)
image.fill(255)
plt.imshow(image[...,::-1]);

traingle= np.array([[200,200],[250,100],[300,200]], np.int32)
pts = traingle.reshape((-1,1,2))
cv2.polylines(image, [traingle], False , (255,200,0), 3)

cv2.rectangle(image, (200,200), (300,300), (255,200,0), -1)

plt.imshow(image[...,::-1]);
```
