```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("/content/drive/MyDrive/Colab Notebooks/Image/adad.jpg",0)


blur = cv2.GaussianBlur(img,(7,7),0)
ret1,output1 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
output2 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,5,2)
output3 = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,5,2)

plt.figure(figsize=[15,7])
plt.subplot(241);plt.imshow(img, cmap='gray');plt.title('Original Noisy Image');
plt.subplot(242);plt.imshow(blur, cmap='gray');plt.title('Gaussian filtered Image');
plt.subplot(245);plt.imshow(output1, cmap='gray');plt.title("Otsu's Thresholding");
plt.subplot(246);plt.imshow(output2, cmap='gray');plt.title("Adaptive Mean + noise removal");
plt.subplot(247);plt.imshow(output3, cmap='gray');plt.title("Adaptive Gaussian + noise removal");
```
