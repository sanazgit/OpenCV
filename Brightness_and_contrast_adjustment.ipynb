```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image= cv2.imread("/content/drive/MyDrive/Colab Notebooks/Image/Weeki_Wachee_spring_10079u.jpg")
image.shape
```

```python
plt.imshow(image[...,::-1]);
```

```python
output_linear1 = cv2.convertScaleAbs(image, alpha=1.3, beta=40)


gamma = [1/4, 1/3, 1/2, 1, 2]
gamma2 = ['1/4', '1/3', '1/2', '1', '2']

output_gamma=[]
ncols = 4
nrows = 2

for g in gamma:
  lookUpTable = np.empty((1,256), np.uint8)
  for i in range(256):
    lookUpTable[0,i] = np.clip(pow(i / 255.0, g) * 255.0, 0, 255)
  output_gamma.append(cv2.LUT(image, lookUpTable))

plt.imshow(output_linear1[...,::-1]);plt.title("Linear");
plt.figure(figsize=[15,8])

for n, pic in enumerate(output_gamma):
  plt.subplot(nrows,ncols, n+1);plt.imshow(output_gamma[n][...,::-1]);plt.title(gamma2[n]);
```
